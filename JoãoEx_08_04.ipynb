1 - Separar números
Faça um programa que:

Leia 10 números;
Crie duas listas:
Uma com números
Outra com números
Depois, mostre as duas listas
🔹 Desafios extras:
Mostrar quantos números tem em cada lista
Mostrar qual lista tem mais números

```python
pares = []
impares = []

for i in range(10):
    n = int(input("Digite um número: "))

    if n % 2 == 0:
        pares.append(n)
    else:
        impares.append(n)

print("Pares:", pares)
print("Ímpares:", impares)

print("Qtd pares:", len(pares))
print("Qtd ímpares:", len(impares))

if len(pares) > len(impares):
    print("Tem mais pares")
elif len(impares) > len(pares):
    print("Tem mais ímpares")
else:
    print("Empate")
```

2 - Procurar número
Faça um programa que:
Tenha uma lista de números (pode ser fixa no código);
Peça um número ao usuário;
Verifique se esse número está na lista;

Resultados:
Se existir → mostrar que encontrou
Se não existir → mostrar que não encontrou
🔹 Desafios extras:
Mostrar a posição onde o número aparece
Mostrar quantas vezes ele aparece

```python
numeros = [1, 3, 5, 7, 3, 9, 3]

n = int(input("Digite um número: "))

if n in numeros:
    print("Achou!")
    print("Posição:", numeros.index(n))
    print("Vezes:", numeros.count(n))
else:
    print("Não achou")
```

3 - Faça um programa que:
Leia 8 números e guarde em uma lista;
Verifique se os números estão em ordem crescente;
Exemplo:

[1, 2, 3, 5, 7] → está em ordem
[1, 3, 2, 5] → não está em ordem

🔹 Desafios extras:
Mostrar em qual posição a ordem está errada
Ordenar a lista manualmente (sem usar sort)

```python
nums = []
for i in range(8):
    nums.append(int(input(f"Número {i+1}: ")))

print("Lista digitada:", nums)

crescente = True
for i in range(7):
    if nums[i] > nums[i+1]:
        crescente = False
        print(f"Erro na posição {i} ({nums[i]} > {nums[i+1]})")
        break

for i in range(8):
    for j in range(7):
        if nums[j] > nums[j+1]:
            nums[j], nums[j+1] = nums[j+1], nums[j]

print("Lista ordenada manualmente:", nums)
```

4 - Sistema bancário simples
Crie um programa com saldo inicial de 1000;
Faça um menu no qual o usuário poderá escolher entre:

1 - Ver saldo
2 - Depositar
3 - Sacar
4 - Sair

Regras:
Não permitir sacar mais do que o saldo
🔹 Desafios extras:
Mostrar histórico das operações
Limitar valor de saque

```python
saldo = 1000
historico = []

while True:
    print("1-Ver saldo 2-Depositar 3-Sacar 4-Sair")
    opc = input("Escolha: ")

    if opc == "1":
        print(f"Saldo: R$ {saldo}")
        print("Histórico:", historico if historico else "Nenhuma operação")

    elif opc == "2":
        v = float(input("Valor: R$ "))
        if v > 0:
            saldo += v
            historico.append(f"Depósito R$ {v}")
        else:
            print("Valor inválido!")

    elif opc == "3":
        v = float(input("Valor: R$ "))
        if v <= 0:
            print("Valor inválido!")
        elif v > saldo:
            print("Saldo insuficiente!")
        elif v > 800:
            print("Limite de saque: R$800")
        else:
            saldo -= v
            historico.append(f"Saque R$ {v}")

    elif opc == "4":
        break

    else:
        print("Opção inválida!")
```

5 - Cadastro de nomes
Faça um programa com menu:

1 - Adicionar nome
2 - Listar nomes
3 - Buscar nome
4 - Remover nome
5 - Sair

Regras:
Usar uma lista para armazenar os nomes
O programa só termina quando escolher “Sair”
🔹 Desafios extras:
Não permitir nomes repetidos
Mostrar quantos nomes estão cadastrados

```python

nomes = []

while True:
    print("1-Adicionar 2-Listar 3-Buscar 4-Remover 5-Sair")
    opc = input("Escolha: ")

    if opc == "1":
        n = input("Nome: ")
        if n not in nomes:
            nomes.append(n)
        else:
            print("Nome já cadastrado!")

    elif opc == "2":
        print("Nomes:", nomes)

    elif opc == "3":
        n = input("Buscar nome: ")
        print("Encontrado!" if n in nomes else "Não encontrado!")

    elif opc == "4":
        n = input("Remover nome: ")
        if n in nomes:
            nomes.remove(n)
        else:
            print("Nome não encontrado!")

    elif opc == "5":
        break

    else:
        print("Opção inválida!")
```
